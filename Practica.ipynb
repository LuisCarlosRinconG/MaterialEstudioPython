```python
#Hacer un diagrama de torta que contenga un titulo y lo valores dividados en la tota (a,b,c)
import matplotlib.pyplot as plt

Valores = [20,20,20]
Etiquetas = ["A","B","C"]

fig,ax=plt.subplots()
ax.pie(Valores, labels=Etiquetas, autopct="%0.1f %%",
      pctdistance=0.4, labeldistance=.6)
ax.set_title("Grafico de pastel")

plt.show()
```
